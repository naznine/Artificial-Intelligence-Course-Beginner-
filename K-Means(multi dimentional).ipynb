```python
import pandas as pd
import numpy as np
import random as rd
import matplotlib.pyplot as plt
import math

dataset=pd.read_csv('H:\RUET\K_Means\yeast1.csv')
#print(dataset)
X = dataset.iloc[:,0:].values
print(X)

total_rows=len(dataset.axes[0]) 
total_cols=len(dataset.axes[1])
str1=dataset.columns[total_cols-1]
#print(str1)
uniq=np.unique(dataset[str1])
print(uniq)

m=X.shape[0] 
n=X.shape[1]
print(m)
print(n)
label=X[:,n-1]
print(label)
K=2
```

```python
centroid1=[]
centroid2=[]
c1=X[0,:]
c2=X[m-1,:]

for i in range(n-1):
    centroid1.append(c1[i])
for i in range(n-1):
    centroid2.append(c2[i])

clushture1=[]
clushture2=[]

for i in range(100):
    print(i)
    print(centroid1)
    print(centroid2)
    it1=1
    it2=1
    for j in range(m):
        rw=X[j,:]
        ds1=0.0
        ds2=0.0
        for j1 in range(n-1):
            ds1+=(centroid1[j1]-rw[j1])*(centroid1[j1]-rw[j1])
        ds1=math.sqrt(ds1)
        for j1 in range(n-1):
            ds2+=(centroid2[j1]-rw[j1])*(centroid2[j1]-rw[j1])
        ds2=math.sqrt(ds2)

        if ds1<ds2:
            clushture1.append(rw)
            it1+=1
        else:
            clushture2.append(rw)
            it2+=1
    #print(clushture1)
    #print(clushture2)

    sm1=[]
    sm2=[]
    for j in range(n-1):
        sm1.append(0)
    for j in range(n-1):
        sm2.append(0)
    for j in range(it1-1):
        tm=clushture1[j]
        for j1 in range(n-1):
            sm1[j1]+=tm[j1]
    for j1 in range(n-1):
        sm1[j1]/=it1-1
    #print(sm1) 

    for j in range(it2-1):
        tm=clushture2[j]
        for j1 in range(n-1):
            sm2[j1]+=tm[j1]
    for j1 in range(n-1):
        sm2[j1]/=it2-1
    #print(sm2)

    f1=0
    f2=0
    for j in range(n-1):
        if centroid1[j]!=sm1[j]:
            f1=1;
    for j in range(n-1):
        if centroid2[j]!=sm2[j]:
            f2=1;
    if f1==0 and f2==0:
        break
    #print(it1)
    #print(it2)
    centroid1=sm1
    centroid2=sm2
    
'''
print("Clushture1:")
for i in range(it1-1):
    print(clushture1[i])
print("Clushture2:")
for i in range(it2-1):
    print(clushture2[i])

'''
```

```python
TN=0
FP=0
FN=0
TP=0

for i in range(it1-1):
    tm=clushture1[i]
    if(tm[n-1]==uniq[0]):
        TP+=1;
    else:
        FP+=1;
for i in range(it2-1):
    tm=clushture2[i]
    if(tm[n-1]==uniq[1]):
        TN+=1;
    else:
        FN+=1;
print('Confusion_Matrix:')
print(TP,' ',FP)
print(FN,' ',TN)

accuracy=((TN+TP)/(TN+FP+TP+FN))*100.0
print('Acuracy=',accuracy,'%')
precision=(TP/(TP+FP))*100.0
print('Pecision=',precision,'%')
recall=(TP/(TP+FN))*100.0
print('Rcall',recall,'%')
F1_Score=2.0*((precision*recall)/(precision+recall))
print('F1_Score=',F1_Score)
```
